```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고가 뜨지 않도록..
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False

# 출력 창 청소
from IPython.display import clear_output

import time


# 평가함수
# 분류용
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

# 회귀용
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# 딥러닝
import tensorflow as tf

# 딥러닝 모델 구조를 정의하는 것
from tensorflow.keras.models import Sequential
# 층 모델(은닉층과 출력층)
from tensorflow.keras.layers import Dense
# 활성화 함수를 정의하는 것
from tensorflow.keras.layers import Activation

# 데이터 표준화
from sklearn.preprocessing import StandardScaler
# 문자열을 숫자로 변환
from sklearn.preprocessing import LabelEncoder
# 원핫인코딩
from tensorflow.keras.utils import to_categorical

# GPU 할당
gpus = tf.config.experimental.list_physical_devices('GPU')
# GPU가 있다면..
if len(gpus) > 0 :
    try :
        for gpu in gpus :
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e :
        print(e)
```

```python
# 랜덤 시드 설정
np.random.seed(1)
tf.random.set_seed(1)
```

```python
# 데이터를 읽어온다.
df1 = pd.read_csv('data/iris.csv', header=None)
df1.columns = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
df1
```

```python
# 데이터 프레임의 정보 확인
df1.info()
```

```python
# 결측치확인
df1.isna().sum()
```

```python
# 입력과 결과로 나눈다.
X = df1.drop('품종', axis=1)
y = df1['품종']

display(X)
display(y)
```

```python
# 문자열 데이터를 숫자로 변환한다.
encoder1 = LabelEncoder()
encoder1.fit(y)
y = encoder1.transform(y)
y
```

```python
# 데이터 표준화
scaler1 = StandardScaler()
scaler1.fit(X)
X = scaler1.transform(X)
X
```

```python
# 원핫 인코딩 수행
# 분류의 경우에는 결과 데이터가 0부터 1씩증가되는 형태라고 하더라도
# 연속성을 가진 데이터가 아닌 각각의 별개의 데이터를 의미한다.
# 딥러닝은 무조건 회귀분석을 수행하기 때문에 원핫 인코딩을 통해
# 연속성을 가지지 않는 형태의 데이터로 만들어 학습 시켜야 한다.
y_encoded = to_categorical(y)
y_encoded
```

```python
# 신경망 구조를 설계한다.
model = Sequential()

a1 = X.shape[1]
model.add(Dense(16, input_dim=a1))
model.add(Activation('relu'))

# 결과 데이터의 종류가 3가지..
model.add(Dense(3))
# 다중 분류이므로....
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
# 손실함수
# 다중 분류이므로 categorical_crossentropy
model.compile(loss='categorical_crossentropy', optimizer='adam', 
              metrics=['accuracy'])
```

```python
# 학습한다.
model.fit(X, y_encoded, epochs=50, batch_size=10)
```

```python
# 성능 확인
a1 = model.evaluate(X, y_encoded)
print(f'손실률 : {a1[0]}')
print(f'정확도 : {a1[1]}')
```

```python
a1 = model.predict(X)
pred = np.argmax(a1, axis=-1)
pred
```

```python
r1 = accuracy_score(y, pred)
r1
```
