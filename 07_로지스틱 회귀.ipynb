```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고가 뜨지 않도록..
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False

# 출력 창 청소
from IPython.display import clear_output

import time
```

```python
# 데이터
x_data = [2, 4, 6, 8, 10, 12, 14]
y_data = [0, 0, 0, 1, 1, 1, 1]
```

```python
# 그래프로 표현하기
plt.scatter(x_data, y_data)
plt.show()
```

```python
# 가중치와 바이어스
a = 0
b = 0
```

```python
# 학습률
lr = 0.05
```

```python
# 시그모이드 함수
def sigmoid(x) :
    return 1 / (1 + np.e ** (-x))
```

```python
# 계산의 편리성을 위해 ndarray를 생성한다.
x1_data = np.array(x_data)
y1_data = np.array(y_data)

print(x1_data)
print(y1_data)
```

```python
# 학습횟수
epochs = 100000
```

```python
# 학습 횟수 만큼 반복한다.
for i in range(epochs) :
    # 현재의 가중치와 바이어스를 이용해 예측 결과를 가져온다.
    # 회귀 분석을 통해 계산된 값을 sigmoid 함수로 다시 계산한다.
    y_pred = sigmoid((a * x1_data) + b)
    # print(y_pred)
    # 오차를 구한다.
    error = y1_data - y_pred
    # print(error)

    # 오차값을 가중치로 미분한 값
    a1_diff = -(2 / len(x1_data)) * sum(x1_data * error)
    # 오차값을 바이어스로 미분한 값
    b_diff = -(2 / len(x1_data)) * sum(error)

    # print(a_diff)
    # print(b_diff)

    # 학습률을 곱해 기존의 가중치와 바이어스를 업데이트 한다.
    a = a - (lr * a1_diff)
    b = b - (lr * b_diff)

    # print(기울기)
    # print(절편)
    
    # 1000 회마다 한번씩 결과를 출력한다.
    if i % 1000 == 0 :
        time.sleep(0.1)
        clear_output(wait=True)
        
        print(f'epoch : {i}, a : {a}, b : {b}')
        
        plt.scatter(x1_data, y1_data)
        
        x_range = np.arange(0, 15, 0.1)
        plt.plot(x_range, np.array([sigmoid(a * x + b) for x in x_range]))
        plt.show()
```

```python
# 결과 예측
pred = sigmoid((a * 16) + b)

if pred > 0.5 : 
    print(1)
else :
    print(0)
```
